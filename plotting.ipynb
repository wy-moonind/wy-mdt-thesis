```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import sys
```

```python
# case 07 data
df = pd.read_csv('../case_21.csv', header=None, sep=',')
df = np.array(df)
print(df.shape)
train_org = df[:5, :7]
val_org = df[:5, 7:14]
r2_org = df[:5, 14:]
print(train_org)
train_org_order = np.average(train_org, axis=0) # 每列平均（不同order）
val_org_order = np.average(val_org, axis=0) # 每列平均（不同order）
r2_org_order = np.average(r2_org, axis=0) # 每列平均（不同order）
train_org_layer = np.average(train_org, axis=1) # 每行平均（不同layer）
val_org_layer = np.average(val_org, axis=1) # 每行平均（不同layer）
r2_org_layer = np.average(r2_org, axis=1) # 每行平均（不同layer）

train_tanh = df[6:11, :7]
val_tanh = df[6:11, 7:14]
r2_tanh = df[6:11, 14:]
train_tanh_order = np.average(train_tanh, axis=0) # 每列平均（不同order）
val_tanh_order = np.average(val_tanh, axis=0) # 每列平均（不同order）
r2_tanh_order = np.average(r2_tanh, axis=0) # 每列平均（不同order）
train_tanh_layer = np.average(train_tanh, axis=1) # 每行平均（不同layer）
val_tanh_layer = np.average(val_tanh, axis=1) # 每行平均（不同layer）
r2_tanh_layer = np.average(r2_tanh, axis=1) # 每行平均（不同layer）

train_relu = df[12, :7]
val_relu = df[12, 7:14]
r2_relu = df[12, 14:]

train_softplus = df[14, :7]
val_softplus = df[14, 7:14]
r2_softplus = df[14, 14:]

layer = [1, 2, 3, 4, 5]
order = [3, 5, 7, 9, 11, 13, 15]
```

```python
# comparing tanh and None
plt.figure(0)
plt.plot(order, train_org_order)
plt.plot(order, train_tanh_order)
plt.grid(True)
plt.ylim(0.15, 0.3)
plt.legend(['None', 'Tanh'])
plt.title('Training loss with/without Tanh activation (different orders)')
plt.xlabel('Order')
plt.ylabel('RMSE loss')
plt.savefig('../figs/activation/multi_train_21_order', dpi=300)

plt.figure(1)
plt.plot(order, val_org_order)
plt.plot(order, val_tanh_order)
plt.grid(True)
plt.ylim(0.15, 0.3)
plt.legend(['None', 'Tanh'])
plt.title('Validation loss with/without Tanh activation (different orders)')
plt.xlabel('Order')
plt.ylabel('RMSE loss')
plt.savefig('../figs/activation/multi_val_21_order', dpi=300)

plt.figure(2)
plt.plot(order, r2_org_order)
plt.plot(order, r2_tanh_order)
plt.grid(True)
plt.ylim(0.3, 0.67)
plt.legend(['None', 'Tanh'])
plt.title('Validation r2 with/without Tanh activation (different orders)')
plt.xlabel('Order')
plt.ylabel('R2')
plt.savefig('../figs/activation/multi_r2_21_order', dpi=300)

plt.figure(3)
plt.plot(layer, train_org_layer)
plt.plot(layer, train_tanh_layer)
plt.grid(True)
plt.ylim(0.15, 0.3)
plt.legend(['None', 'Tanh'])
plt.title('Train loss with/without Tanh activation (different layers)')
plt.xlabel('Layer')
plt.ylabel('RMSE loss')
plt.savefig('../figs/activation/multi_train_21_layer', dpi=300)

plt.figure(4)
plt.plot(layer, val_org_layer)
plt.plot(layer, val_tanh_layer)
plt.grid(True)
plt.ylim(0.15, 0.3)
plt.legend(['None', 'Tanh'])
plt.title('Validation loss with/without Tanh activation (different layers)')
plt.xlabel('Layer')
plt.ylabel('RMSE loss')
plt.savefig('../figs/activation/multi_val_21_layer', dpi=300)

plt.figure(5)
plt.plot(layer, r2_org_layer)
plt.plot(layer, r2_tanh_layer)
plt.grid(True)
plt.ylim(0.3, 0.67)
plt.legend(['None', 'Tanh'])
plt.title('Validation r2 with/without Tanh activation (different layers)')
plt.xlabel('Layer')
plt.ylabel('R2')
plt.savefig('../figs/activation/multi_r2_21_layer', dpi=300)
```

```python
# case 07 tanh single
plt.figure(0)
# train loss
plt.plot(order, train_org[0,:])
plt.plot(order, train_tanh[0,:])
plt.plot(order, train_relu[:])
plt.plot(order, train_softplus[:])
plt.legend(['None', 'Tanh', 'ReLU', 'Softplus'])
plt.ylim(0.1, 0.4)
plt.title('Different activation for single layer (train loss)')
plt.xlabel('Order')
plt.ylabel('RMSE loss')
plt.grid(True)
plt.savefig('../figs/activation/single_train_21', dpi=300)

plt.figure(1)
# val loss
plt.plot(order, val_org[0,:])
plt.plot(order, val_tanh[0,:])
plt.plot(order, val_relu[:])
plt.plot(order, val_softplus[:])
plt.legend(['None', 'Tanh', 'ReLU', 'Softplus'])
plt.ylim(0.1, 0.4)
plt.title('Different activation for single layer (validation loss)')
plt.xlabel('Order')
plt.ylabel('RMSE loss')
plt.grid(True)
plt.savefig('../figs/activation/single_val_21', dpi=300)

plt.figure(2)
# val r2
plt.plot(order, r2_org[0,:])
plt.plot(order, r2_tanh[0,:])
plt.plot(order, r2_relu[:])
plt.plot(order, r2_softplus[:])
plt.legend(['None', 'Tanh', 'ReLU', 'Softplus'])
plt.ylim(0, 0.7)
plt.title('Different activation for single layer (validation R2)')
plt.xlabel('Order')
plt.ylabel('R2 value')
plt.grid(True)
plt.savefig('../figs/activation/single_r2_21', dpi=300)
```
